# Δημιουργία αρχείου year.

```python
import io
year = [i+1960 for i in range(60)]
five_year_period = ["1960-1964","1965-1969","1970-1974","1975-1979","1980-1984","1985-1989","1990-1994","1995-1999","2000-2004","2005-2009","2010-2014","2015-2019"]
ten_year_period = ["1960-1969","1970-1979","1980-1989","1990-1999","2000-2009","2010-2019"]
twenty_year_period = ["1960-1979","1980-1999","2000-2019"]
year
```

```python
five=[]
ten =[]
twenty=[]
for i in range(60):
    if(i+1960 >=2000 and i+1960 <=2019):
        twenty.append("2000-2019")
        if(i+1960>=2000 and i+1960<=2009):
            ten.append("2000-2009")
            if(i+1960>=2000 and i+1960<=2004):
                five.append("2000-2004")
            elif(i+1960>=2005 and i+1960<=2009):
                five.append("2005-2009")
        if(i+1960>=2010 and i+1960<=2019):
            ten.append("2010-2019")
            if(i+1960>=2010 and i+1960<=2014):
                five.append("2010-2014")
            elif(i+1960>=2015 and i+1960<=2019):
                five.append("2015-2019")
    elif(i+1960 >=1980 and i+1960 <=1999):
        twenty.append("1980-1999")
        if(i+1960>=1980 and i+1960<=1989):
            ten.append("1980-1989")
            if(i+1960>=1980 and i+1960<=1984):
                five.append("1980-1984")
            elif(i+1960>=1985 and i+1960<=1989):
                five.append("1985-1989")
        if(i+1960>=1990 and i+1960<=1999):
            ten.append("1990-1999")
            if(i+1960>=1990 and i+1960<=1994):
                five.append("1990-1994")
            elif(i+1960>=1995 and i+1960<=1999):
                five.append("1995-1999")
    elif(i+1960 >=1960 and i+1960 <=1979):
        twenty.append("1960-1979")
        if(i+1960>=1960 and i+1960<=1969):
            ten.append("1960-1969")
            if(i+1960>=1960 and i+1960<=1964):
                five.append("1960-1964")
            elif(i+1960>=1965 and i+1960<=1969):
                five.append("1965-1969")
        if(i+1960>=1970 and i+1960<=1979):
            ten.append("1970-1979")
            if(i+1960>=1970 and i+1960<=1974):
                five.append("1970-1974")
            elif(i+1960>=1975 and i+1960<=1979):
                five.append("1975-1979")
```

```python
import pandas as pd
data= {"year":year, "five":five,"ten":ten,"twenty":twenty}
df = pd.DataFrame(data=data)
df
```
